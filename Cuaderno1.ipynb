### Es primo verificará si el número es primo o no

````python
def esprimo(num):
    div = 2
    while(div < num):
        if(num % div == 0):
            return False
        div += 1
    return False 
````

### Sumatoria

````python
def sumatoria(i,n):
    resultado = 0
    for numero in range(i,n+1):
        resultado += numero
    return resultado 
````
